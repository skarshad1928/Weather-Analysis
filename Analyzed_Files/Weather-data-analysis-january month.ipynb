```python
import requests
import pandas as pd

url = "https://archive-api.open-meteo.com/v1/archive"

latitude = 32.8546
longitude = -79.9748

years = [2022, 2023, 2024, 2025]

all_data = []

for year in years:
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": f"{year}-01-01",
        "end_date": f"{year}-01-31",
        "hourly": "temperature_2m,relative_humidity_2m",
        "timezone": "auto"
    }

    response = requests.get(url, params=params)
    data = response.json()

    df = pd.DataFrame({
        "time": data["hourly"]["time"],
        "temperature": data["hourly"]["temperature_2m"],
        "humidity": data["hourly"]["relative_humidity_2m"]
    })

    df["time"] = pd.to_datetime(df["time"])

    df = df[df["time"].dt.hour == 12]

    df["year"] = year

    all_data.append(df)

final_df = pd.concat(all_data, ignore_index=True)


final_df["date"] = final_df["time"].dt.date
final_df = final_df.drop(columns=["time"])

final_df.head()
```

```python
import requests
import pandas as pd

url = "https://archive-api.open-meteo.com/v1/archive"

latitude = 32.8546
longitude = -79.9748

years = [2022, 2023, 2024, 2025]

all_data = []

for year in years:
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": f"{year}-01-01",
        "end_date": f"{year}-01-31",
        "hourly": "precipitation",
        "timezone": "auto"
    }

    response = requests.get(url, params=params)
    data = response.json()

    df = pd.DataFrame({
        "time": data["hourly"]["time"],
        "precipitation": data["hourly"]["precipitation"]
    })
    df["time"] = pd.to_datetime(df["time"])
    df = df[df["time"].dt.hour == 12]
    df["year"] = year
    df["date"] = df["time"].dt.date

    all_data.append(df)


final_precip_df = pd.concat(all_data, ignore_index=True)

final_precip_df.head()
```

```python
all_wind = []

for year in years:
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "start_date": f"{year}-01-01",
        "end_date": f"{year}-01-31",
        "hourly": "winddirection_10m",
        "timezone": "auto"
    }

    response = requests.get(url, params=params)
    data = response.json()

    df_wind = pd.DataFrame({
        "time": data["hourly"]["time"],
        "wind_direction": data["hourly"]["winddirection_10m"]
    })

    df_wind["time"] = pd.to_datetime(df_wind["time"])

    # Keep only 12:00 PM
    df_wind = df_wind[df_wind["time"].dt.hour == 12]

    df_wind["year"] = year
    df_wind["date"] = df_wind["time"].dt.date

    all_wind.append(df_wind)

wind_df = pd.concat(all_wind, ignore_index=True)
```

```python
def degree_to_compass(deg):
    if pd.isna(deg):
        return None
    directions = ["N", "NE", "E", "SE", "S", "SW", "W", "NW"]
    idx = int((deg + 22.5) // 45) % 8
    return directions[idx]
```

```python
final_df["wind_direction"] = final_df["wind_direction"].apply(degree_to_compass)
```

Rain can occur due to three to four major meteorological reasons, especially when large-scale atmospheric and oceanic conditions change. One important reason is the influence of cyclonic systems formed over the North Atlantic Ocean, which disturb normal wind and pressure patterns. During a cyclone, low-pressure regions are created, causing moist air to rise, cool, and condense into clouds, eventually leading to rainfall.

Another major factor is the interaction of north-west and south-west wind systems, which transport moisture from oceanic regions toward land. When these moist winds converge or are forced to rise due to pressure differences, rainfall becomes more likely. Additionally, changes in river water levels and surrounding surface moisture can indicate increased evaporation and atmospheric instability, which further supports cloud formation. In regions influenced by cyclonic activity, the combined effect of wind convergence, moisture availability, pressure drop, and oceanâatmosphere interaction leads to enhanced rainfall. Thus, cyclones originating in the North Atlantic Ocean, along with wind circulation and hydrological factors, play a significant role in triggering rainfall in affected regions.
